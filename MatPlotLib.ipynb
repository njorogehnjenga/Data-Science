# Bar chat

```python
#import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Vertical bar chats

```python
plt.figure(figsize=(8,5), dpi=100)
#indicate the data you want to show
modules= ['Python','C++','Java','Perl','Scala','Lisp']
scores= [75,81,69,55,63,52]

#specify positions of the data labels
position= np.arange(len(modules))
#position= [0,1,2,3,4,5]

#plot the bar
plt.bar(position, scores, color='red')

#show appropriate labels
plt.xticks(position,modules)

#add title
plt.title('Programming language usage')

#label the axes
plt.xlabel('Module')
plt.ylabel('Average Score')

#show plot
plt.show()
```

## Horizontal Bar chat

```python
plt.figure(figsize=(8,5), dpi=100)
#indicate the data you want to show
modules= ['Python','C++','Java','Perl','Scala','Lisp']
scores= [75,81,69,55,63,52]

#specify positions of the data labels
#position= np.arange(len(modules))
position= [0,1,2,3,4,5]

#plot the bar
plt.barh(position, scores, color='red')

#show appropriate labels
plt.yticks(position,modules)

#add title
plt.title('Programming language usage')

#label the axes
plt.xlabel('Module')
plt.ylabel('Average Score')

#show plot
plt.show()
```

## Line Graph

```python
gas_df= pd.read_csv('gas_prices.csv')
gas_df
```

```python
plt.figure(figsize=(10,6),dpi=100)
x= gas_df.Year
y1= gas_df.Australia
y2= gas_df.Italy
y3= gas_df.France
plt.plot(x,y1, 'b.-', label='Australia')
plt.plot(x,y2, 'r.-', label='Italy')
plt.plot(x,y3, 'g.-', label='France')
plt.legend()
plt.show()

# for loop can also be used
# countries_to_look_at= ['Australia','Italy','France']
# for countries in countries_to_look_at:
#      if countries in countries_to_look_at:
#        plt.plot(gas_df.Year, gas_df[country], marker='.')
    
```

```python
plt.figure(figsize=(10,6),dpi=100)
countries_to_look_at= ['Australia','Italy','France']
for countries in countries_to_look_at:
    if countries in countries_to_look_at:
        plt.plot(gas_df.Year, gas_df[countries], marker='.', label= countries)
plt.xticks(gas_df['Year'][::3])
plt.title('Gas Prices', fontsize=50)
plt.xlabel('Year')
plt.ylabel('Gas Prices (in USD)')
plt.legend()
plt.show()
```

```python
fifa_df= pd.read_csv('fifa_data.csv')
fifa_df
```

# Histograms

```python
#set the size
plt.figure(figsize=(10,6),dpi=100)

#set bins
bins=[30,40,50,60,70,80,90,100]

#plot histogram
plt.hist(fifa_df['Overall'],bins=bins, density=True)
plt.xticks(bins)

plt.title('Player skill level')

plt.xlabel('Skill level')
plt.ylabel('Players')
plt.show()
```

# Pie charts

```python
# count the number of people that prefer left to right
left= fifa_df.loc[fifa_df['Preferred Foot'] =='Left'].count()[0]
right= fifa_df.loc[fifa_df['Preferred Foot'] =='Right'].count()[0]

#add labels
labels=['left','right']

#add colors
colors=['red','blue']

#plot chart
plt.pie([left,right], labels= labels, colors=colors, autopct='%.2f%%')

plt.title('Preferred Foot')

plt.show()
```
